## Population and Sample parameters

Sample Variance is almost always less than Population Variance unless the Population mean is exactly same as the Sample Mean


1. We can prove it by taking the derivative of the variance and equate it to zero. we identified that the sample mean is equavalent to the minimum variance.
2. We tried it with different mean values and found it this

3. Dividing by (n-1) for the sample variance compensates for the fact that the we are calculating differences from the sample mean instead of the population mean

4. While calculating the variance, it is easy if you can take the squared difference as it is easy to calculate the minimum using derivates without sharp edges which is the case for absolute value



```python
def calculate_mean_variance(arr):
  arr_mean= round(np.mean(arr),2)
  arr_var = round(np.var(arr),2)
  # print(f'Mean :{arr_mean} and Standard Deviation is :{arr_std}')
  return arr_mean,arr_var
```

```python
pop = np.random.randint(1,100,10000)
```

```python
pop_mean,pop_var  = calculate_mean_variance(pop)
print(pop_var)
```

```python
count=0
ls=[]
for i in range(100000):
  sample = np.random.choice(pop,size=10)
  sample_mean,sample_var = calculate_mean_variance(sample)
  ls.append(sample_var)
  if sample_var>pop_var:
    count+=1
  else:
    continue
print(f'Count: ',count)
```

```python
ax = sns.distplot(ls)
x = ax.lines[0].get_xdata() # Get the x data of the distribution
y = ax.lines[0].get_ydata() # Get the y data of the distribution
maxid = np.max(y) # The id of the peak (maximum of y data)
plt.plot(pop_var,maxid,'go--',linewidth=10, ms=10)
plt.show()
```
